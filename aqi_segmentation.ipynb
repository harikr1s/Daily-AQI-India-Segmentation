```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('city_day.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# changing 'Date' attribute to datetime

df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df.set_index('Date', inplace = True)
df.head()
```

```python
# missing values

df.isnull().sum()
```

```python
# missing values by percentage

(100*df.isnull().sum()/len(df)).sort_values(ascending = False)
```

```python
df['City'].unique()
```

```python
df['AQI_Bucket'].unique()
```

```python
df.dropna(how = 'all', inplace = True)
df.shape
```

```python
df['AQI_Bucket'].fillna('Moderate', inplace = True)
for i in ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI']:
  df[i].fillna(df[i].mean(), inplace = True)
```

```python
df['BTX'] = df['Benzene'] + df['Toluene'] + df['Xylene']
df['Particulates'] = df['PM2.5'] + df['PM10']
df.drop(['Benzene', 'Toluene', 'Xylene', 'PM2.5', 'PM10', 'NO','NO2'], axis = 1,inplace = True)
df.head()
```

```python
df_c = df.copy()
df_c.drop(['City', 'AQI_Bucket'], axis = 1, inplace = True)
df_c.corr()
plt.figure(figsize = (15, 8))
sb.heatmap(df_c.corr(), annot = True)
plt.savefig('correlation_heatmap.png')
```

AQI's correlation is high with CO and Particulates

```python
df.index.min().date()
```

```python
df.index.max().date()
```

```python
df[['City', 'AQI']].groupby('City').mean().sort_values('AQI').plot(kind = 'barh', figsize = (16, 8))
plt.title('Average Air Quality Index over 2015-2020 by Cities')
plt.savefig('aqi_by_city.png')
```

Ahmedabad has the worst Air Quality Index, followed by Delhi, Patna, Gurugram and Lucknow.

```python
pollutants = ['NOx', 'NH3', 'CO', 'SO2', 'O3', 'BTX', 'Particulates']
df[pollutants].plot(kind = 'line', figsize=(18,18), subplots = True)
plt.title('Distribution of various pollutants over 2015-2020')
plt.savefig('pollutants_over_the_years.png')
```

SO2 levels were comparatively lower from  mid-2015 to 2018.

BTX levels were higher than usual for a couple of months in 2020.

```python
from IPython.display import display_html 

print('Cities Leaderboard by Pollutants')

df_1 = df[['NOx', 'City']].groupby(["City"]).mean().sort_values(by = 'NOx', ascending = False)[:10]
df_2 = df[['NH3', 'City']].groupby(["City"]).mean().sort_values(by = 'NH3', ascending = False)[:10]
df_3 = df[['CO', 'City']].groupby(["City"]).mean().sort_values(by = 'CO', ascending = False)[:10]
df_4 = df[['SO2', 'City']].groupby(["City"]).mean().sort_values(by = 'SO2', ascending = False)[:10]
df_5 = df[['O3', 'City']].groupby(["City"]).mean().sort_values(by = 'O3', ascending = False)[:10]
df_6 = df[['BTX', 'City']].groupby(["City"]).mean().sort_values(by = 'BTX', ascending = False)[:10]
df_7 = df[['Particulates', 'City']].groupby(["City"]).mean().sort_values(by = 'Particulates', ascending = False)[:10]

df1 = df_1.style.set_table_attributes("style = 'display:inline'")
df2 = df_2.style.set_table_attributes("style = 'display:inline'")
df3 = df_3.style.set_table_attributes("style = 'display:inline'")
df4 = df_4.style.set_table_attributes("style = 'display:inline'")
df5 = df_5.style.set_table_attributes("style = 'display:inline'")
df6 = df_6.style.set_table_attributes("style = 'display:inline'")
df7 = df_7.style.set_table_attributes("style = 'display:inline'")
    
display_html(df1._repr_html_() + "\xa0" + 
             df2._repr_html_() + "\xa0" + 
             df3._repr_html_() + "\xa0" + 
             df4._repr_html_() + "\xa0" + 
             df5._repr_html_() + "\xa0" + 
             df6._repr_html_() + "\xa0" + 
             df7._repr_html_(), 
             raw = True)
```

Ahmedabad, Delhi and Patna are within the top 10 for all pollutant categories.

Ahmedabad has the highest amount of CO and SO2 presence in the air, while the air in Delhi has the highest amount of particulates.
